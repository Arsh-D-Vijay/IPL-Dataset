```python
# Top choice made by teams after winning a toss

fig,ax = plt.subplots(figsize=(15,8))
sns.countplot(x='toss_winner',data = mat_df,hue='toss_decision')
plt.xlabel('Team Name')
plt.ylabel('Toss Won')
plt.title('Toss Winner Decision')
plt.xticks(rotation='vertical')

plt.show()
```

###### Insight: 
Apart from CSK, Deccan Chargers and Pune Warriors all teams prefered to opt for Fielding first.

```python
# Analyzing if the decision made by team was right or not 

## Creating Derived Columns

mat = mat_df.dropna(subset = ['winner'],axis =0)
mat['right_de'] = np.asarray(mat.toss_winner == mat.winner)

## Plotting

fig,ax = plt.subplots(figsize=(15,8))
sns.countplot(x='toss_winner',data = mat,hue='right_de')
plt.xlabel('Team Name')
plt.ylabel('Wins')
plt.title('Toss Winner Decision Effect on Winning')
plt.xticks(rotation='vertical')
plt.legend(labels = ['Wrong Decision','Right Decision'])

plt.show()
```

##### Insight: 
CSK has made the right call most of times whenever they won a toss. Whereas Pune Warriors had most made wrong decision which led the team to lose the match.

```python
# Top 10 Hiters of the series


# Some players had played only few balls so it would be unfair to plot the graph directly so we will make a /
# /threshold 0f 500 balls to plot the graph

runs_df.sort_values(by=['strikerate'],ascending = False,inplace = True)
Player = runs_df.batsman[runs_df.numberofballs>500].head(10)
SR = runs_df.strikerate[runs_df.numberofballs>500].head(10) # minimum 500 balls has been played by the player

# plotting

fig,ax = plt.subplots(figsize=(15,8))
sns.lineplot(Player,SR,marker = 'o',color = 'r')
sns.barplot(Player,SR, palette = 'pastel')

for index,data in enumerate(SR):
    plt.text(x=index-0.18, y =data+3, s="{:.2f}".format(data) , fontdict=dict(fontsize=12))
plt.xlabel('Player Name')
plt.ylabel('Strike Rate')
plt.title('Top 10 Hiters')
plt.show()
```

###### Insight:
AD Russell has a great record strike rate of 188.16.

```python
# Top 10 highest Individual Score.


indi = deli_df.groupby(['match_id','batsman',])['batsman_runs'].sum().reset_index()
indi.sort_values('batsman_runs',axis=0, inplace=True,ascending=False)
indi = indi[:10]

indi.drop('match_id',inplace=True,axis=1)
indi.set_index('batsman',inplace=True)

#Plotting 
indi.plot(kind= 'bar' , figsize = (15,8), color = green)
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Top 10 Individual Scores')
plt.show()
```

##### Insight:
1. Chris Gayle and AB de Villiers has been in list twice (as expected from their striking rate graph)

```python
# Orange Cap Holders

name,runs = [],[]
for season in mat_df['Season'].unique():
    new_df = mat_df[mat_df['Season']==season]
    temp_df = deli_df[deli_df['match_id'].isin(new_df['id'].values)]
    temp_df = temp_df.groupby('batsman')['batsman_runs'].sum().reset_index()
    temp_df.sort_values('batsman_runs', inplace=True,ascending=False)
    temp_df = temp_df.iloc[0,:]
    name.append(temp_df['batsman'])
    runs.append(temp_df['batsman_runs'])

orange_df = pd.DataFrame(data={'Season':mat_df['Season'].unique(), 'Player':name, 'Total Runs':runs}, index = None)
print(orange_df.sort_values('Season'))
```

###### Insight: 
1. DA Warner has been Orange Cap Holder 3 times.
2. Virat Kholi had the maximum runs ever scored by the orange cap holder in a season

## Bowlers Analysis

```python
players_df.head()
```

```python
players_df.Bowling_Skill.nunique()
```

```python
# Bowling Skill 

fig,ax = plt.subplots(figsize=(15,8))
sns.countplot(y='Bowling_Skill',data = players_df)
plt.ylabel('Bowling Style')
plt.xlabel('Count')
# plt.title('Battin')

plt.show()
```

```python
# Purple Cap Holders

name,wickets = [],[]

out = ['caught', 'bowled', 'lbw', 'caught and bowled',  'stumped', 'hit wicket']
for season in mat_df['Season'].unique():
    season_df = mat_df[mat_df['Season']==season]
    
    temp_df = deli_df[deli_df['match_id'].isin(season_df['id'].values)]
    temp_df = temp_df[temp_df['dismissal_kind'].isin(out)]
    temp_df = temp_df.groupby('bowler')['dismissal_kind'].count().reset_index()
    
    temp_df.sort_values('dismissal_kind', inplace=True,ascending=False)
    temp_df = temp_df.iloc[0,:]
   
    name.append(temp_df['bowler'])
    wickets.append(temp_df['dismissal_kind'])

purple_df = pd.DataFrame(data={'Season':mat_df['Season'].unique(), 'Bowler':name, 'Wickets':wickets})
purple_df.sort_values('Season')
```

```python
# Most Common Out Type:
plt.figure(figsize=(15,8))
sns.countplot(data=deli_df, x='dismissal_kind')
plt.xticks(rotation=90)
plt.title('Common Out Type')
plt.show()
```

###### Insight:
1. The most common type of out is Catch Out
2. There were cases where player was assigned Out due to Obstructing the Field, injury or Hit wicket but as expected these are rare cases

```python
# Top 10 Wicket Takers in Overall Series
out = ['caught', 'bowled', 'lbw', 'caught and bowled',  'stumped', 'hit wicket']
wickets_taker = deli_df[deli_df['dismissal_kind'].isin(out)]
wickets_taker = wickets_taker.groupby('bowler')['dismissal_kind'].count().reset_index()
wickets_taker.sort_values('dismissal_kind', ascending=False, inplace=True)
wickets_taker.set_index('bowler', inplace=True)
wickets_taker.head(10)
wickets_taker[:10].plot(kind= 'bar' , figsize = (15,8), color = crimson)
plt.legend(labels = ['Wicket'])
plt.show()
```

###### Insight:
1. Malinga has undisputed lead in taking most wickets 

```python
# Top 10 bowlers with most dot balls

dot_ball = deli_df[deli_df['total_runs']==0]
dot_ball = dot_ball['bowler'].value_counts()[:10]
dot_ball.plot(kind='bar', figsize=(11,6), title='Bowlers who have maximum number of Dot balls')

plt.xlabel('BOWLER')
plt.ylabel('BALLS')
plt.show()
```
